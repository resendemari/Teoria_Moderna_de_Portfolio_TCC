## Selecionando os fundos pelo CNPJ e pivotando a tabela para ter os fundos nas colunas



```python
filt = conteudo1['CNPJ_FUNDO'].isin(cnpj)
fundos_escolhidos = conteudo1.loc[filt]
precos_fundos = fundos_escolhidos.pivot(index='DT_COMPTC', columns='CNPJ_FUNDO',values=['VL_QUOTA'])
```

```python
carteira = pd.DataFrame()
for i in acoes:
  carteira[i] = yf.download(i, period='5y')['Adj Close']

precos_acoes = pd.DataFrame(carteira)
```

#An√°lise de t√≠tulos

## Calculando os retornos em fun√ß√£o do pre√ßo anterior

```python
# Estimar os retornos di√°rios
retornos_fundos = precos_fundos.pct_change()
retornos_acoes = precos_acoes.pct_change() 
```

## Unindo os dados de a√ß√µes e fundos em um dataframe

```python
d_acoes = {}
d_fundos = {}

for i in range(len(acoes)):
  d_acoes[acoes[i]]=retornos_acoes[acoes[i]]

for i in range(len(cnpj)):
  d_fundos[cnpj[i]]=retornos_fundos['VL_QUOTA'][cnpj[i]]

if (cnpj):
  dataframe = retornos_fundos['VL_QUOTA'].index
  retornos_fundos_df = pd.DataFrame(index=np.array(dataframe),data = d_fundos)
  retornos_fundos_df.index = pd.to_datetime(retornos_fundos_df.index)
else:
  dataframe = retornos_fundos
  retornos_fundos_df = pd.DataFrame(dataframe)
  

if (acoes): 
  dataframe = retornos_acoes.index
  retornos_acoes_df = pd.DataFrame(index=np.array(dataframe),data = d_acoes)
  retornos_acoes_df.index = pd.to_datetime(retornos_acoes_df.index) 
else:
  dataframe = retornos_fundos
  retornos_acoes_df = pd.DataFrame(dataframe)

if(len(acoes) & len(cnpj)):
  retornos_df = pd.merge(retornos_acoes_df,retornos_fundos_df,left_index=True, right_index=True)
elif(not(len(cnpj)) & len(acoes)):
  retornos_df = retornos_acoes_df
elif(not(len(acoes)) & cnpj):
  retornos_df = retornos_fundos_df



```

## C√°lculo do retorno esperado, do desvio-padr√£o e da matriz de covari√¢ncia anualizados

```python
e_r = retornos_df.mean()   #retorno esperado
var = retornos_df.var()    #variancia
vol = var**0.5             #volatilidade = desvio-padrao
matriz_cov = retornos_df.cov()  #matriz de covariancia

e_r = e_r * 252
matriz_cov = matriz_cov * 252
vol = var*np.sqrt(252)
```

#Simula√ß√£o de carteiras

```python
noa = len(retornos_df.columns)  #numero de ativos na carteira
nop = 50000                    #numero de portfolios simulados

p_ret = []
p_vol = []
p_pesos = []

for portfolios in range(nop):
  pesos = np.random.random(noa)
  pesos = pesos/np.sum(pesos)
  p_pesos.append(pesos)
  p_ret.append(np.sum(e_r*pesos))
  p_vol.append(np.sqrt(np.dot(pesos.T,np.dot(matriz_cov,pesos))))


p_ret = np.array(p_ret)
p_vol = np.array(p_vol)
dados = {'Retorno': p_ret,'Risco':p_vol}
portfolios = pd.DataFrame(dados)
portfolios['Sharpe'] = (portfolios['Retorno']-rf)/portfolios['Risco']


min_risk = portfolios[portfolios['Risco']==portfolios['Risco'].min()]
max_sharpe = portfolios[portfolios['Sharpe']==portfolios['Sharpe'].max()]
indice_min_risk = min_risk.index.values.astype(int)[0]
indice_max_sharpe = max_sharpe.index.values.astype(int)[0]
```

##Plot carteiras simuladas


```python
plt.figure(figsize=(10, 6))
plt.scatter(p_vol, p_ret, c=(p_ret-rf) / p_vol, marker='o',alpha=0.8, cmap='plasma')
plt.grid(True)
plt.xlabel('Volatilidade',fontsize=label_size)
plt.ylabel('Retorno esperado',fontsize=label_size)
plt.plot(min_risk['Risco'],min_risk['Retorno'], 'go', markersize=10.0)
plt.plot(max_sharpe['Risco'],max_sharpe['Retorno'], 'co', markersize=10.0)
cb = plt.colorbar()
cb.set_label(label='Indice de Sharpe',size=label_size)
cb.ax.tick_params(labelsize=ticks_size)
plt.xticks(fontsize=ticks_size)
plt.yticks(fontsize=ticks_size)


plt.savefig('carteiras_simuladas.pdf',format='pdf',dpi=300,transparent=False,bbox_inches='tight')
```

#Otimiza√ß√£o

```python
#Definindo funcoes para calcular a minima variancia e o maximo indice de sharpe 

def statistics(pesos):
  pesos = np.array(pesos)
  pret = np.sum(e_r*pesos)
  pvol = math.sqrt(np.dot(pesos.T,np.dot(matriz_cov,pesos)))
  return np.array([pret,pvol,(pret-rf)/pvol])

def min_func_variance(pesos):
  return statistics(pesos)[1]**2
  #return statistics(pesos)[1]

def max_func_sharpe(pesos):
  return -statistics(pesos)[2]

def min_func_port(pesos):
  return statistics(pesos)[1]


pesos_i = np.array(noa*[1/noa,]) #condicao inicial
bnds = tuple((0, 1) for x in range(noa)) #limites
cons =  ({'type':'eq','fun':lambda x:np.sum(x)-1}) #restricao


otim_vol = sco.minimize(min_func_variance,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)
otim_sharpe = sco.minimize(max_func_sharpe,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)


```

##Print resultados da otimiza√ß√£o

```python
print('\n\n-------------- Carteira M√≠nimo Risco Global -----------\n')

print('Retorno:',round(statistics(otim_vol['x'])[0]*100,2),'%')
print('Risco:',round(statistics(otim_vol['x'])[1]*100,2),'%')
print('Sharpe:',statistics(otim_vol['x']).round(2)[2])
print('Pesos da carteira com minima volatilidade: ',otim_vol['x'].round(3))

print('\n\n-------------- Carteira M√°ximo Sharpe -----------\n')

print('Retorno:',round(statistics(otim_sharpe['x'])[0]*100,2),'%')
print('Risco:',round(statistics(otim_sharpe['x'])[1]*100,2),'%')
print('Sharpe:',statistics(otim_sharpe['x']).round(2)[2])
print('Pesos da carteira com maximo sharpe: ',otim_sharpe['x'].round(3))
```

##Fronteira de M√≠nima Vari√¢ncia

```python
def min_func_ret(pesos):
  return statistics(pesos)[0]

def max_func_ret(pesos):
  return -statistics(pesos)[0]

min_ret = sco.minimize(min_func_ret,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)
max_ret = sco.minimize(max_func_ret,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)

pesos_i = np.array(noa*[1/noa,]) #condicao inicial
bnds = tuple((0, 1) for x in range(noa)) #limites
cons =  ({'type':'eq','fun':lambda x: statistics(x)[0]-tret},
         {'type':'eq','fun':lambda x:np.sum(x)-1}) #restricoes


trets = np.linspace(min_ret['fun'],-max_ret['fun'],50)
tvols=[]

for tret in trets:
  res = sco.minimize(min_func_port,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)
  tvols.append(res['fun'])

tvols = np.array(tvols)
```

#Limites de Aloca√ß√£o

##Definindo limites de aloca√ß√£o para cada ativo (bounds)


```python
limite_w_acoes = (0.01,0.15)
limite_w_fundos = (0.02,0.25)

dict_ativos = {}
dict_acoes = {}
dict_fundos = {}
ativos = np.array(acoes+cnpj)

if((len(acoes)) & len(cnpj)):

  for i in range(len(acoes)):
    dict_ativos[i]=[acoes[i],'acao']
  for j in range(i+1,len(ativos)):
    dict_ativos[j]=[cnpj[j-len(acoes)],'fundo']

elif(not(len(acoes))):
    for i in range(len(cnpj)):
      dict_ativos[i]=[cnpj[i],'fundo']

elif(not(len(cnpj))):
    for i in range(len(acoes)):
      dict_ativos[i]=[acoes[i],'acao']

limites = []
for i in dict_ativos:
  if(dict_ativos[i][1] == 'acao'):
    limites.append(limite_w_acoes)
  elif(dict_ativos[i][1] == 'fundo'):
    limites.append(limite_w_fundos)

bnds = tuple(limites)
```

##Otimiza√ß√£o com limites de aloca√ß√£o

```python
pesos_i = np.array(noa*[1/noa,]) #condicao inicial
cons =  ({'type':'eq','fun':lambda x:np.sum(x)-1}) #restricao

otim_vol_l = sco.minimize(min_func_variance,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)
otim_sharpe_l = sco.minimize(max_func_sharpe,pesos_i,method='SLSQP',bounds=bnds,constraints=cons)

print('\n\n-------------- Carteira M√°ximo Sharpe -----------\n')
print('Retorno:',round(statistics(otim_sharpe_l['x'])[0]*100,2),'%')
print('Risco:',round(statistics(otim_sharpe_l['x'])[1]*100,2),'%')
print('Sharpe:',statistics(otim_sharpe_l['x']).round(2)[2])
print('Pesos da carteira com maximo sharpe: ',otim_sharpe_l['x'].round(3))
```
